```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_movies = pd.read_csv("C:\\Users\\Geksmode\\Desktop\\data\\tmdb_5000_movies.csv")
df_credits = pd.read_csv("C:\\Users\\Geksmode\\Desktop\\data\\tmdb_5000_credits.csv")
```

```python
df_credits.columns = ['id','tittle','cast','crew']
df_movies= df_movies.merge(df_credits,on='id')
```

```python
df_movies
```

```python
#Rating = ((V/(v+m))*R)+(m/(v+m)*C) avec v le nombre de vote pour le film, m le nombre minimum de vote requis pour être dans le chart, R la moyenne des notes du film, et C la moyenne des votes par rapport a tout le chart

C = df_movies["vote_average"].mean()
C
```

```python
#pour m on choisi le nombre de vote minimum tel que les films doivent avoir Il doit avoir plus de votes qu'au moins 90% des films de la liste.

m = df_movies['vote_count'].quantile(0.9)
m
```

```python
qualify_movies = df_movies.copy().loc[df_movies["vote_count"]>=m]
qualify_movies.shape
```

```python
def rating (x,m=m,C=C):
    v = x['vote_count']
    R = x['vote_average']
    return ((v/(v+m))*R)+(m/(v+m)*C)
    
```

```python
qualify_movies['score'] = qualify_movies.apply(rating, axis=1)
```

```python
#trie les movies basé sur le score calculer :

qualify_movies = qualify_movies.sort_values('score',ascending=False)

qualify_movies[['title','vote_count','vote_average','score','popularity']].head(10)
```

```python
#define the popularity of the movie

pop = df_movies.sort_values("popularity",ascending = False)
plt.figure(figsize=(12,4))
plt.barh(pop['title'].head(6),pop['popularity'].head(6),align = 'center',color = 'blue')
plt.gca().invert_yaxis()
plt.xlabel("Popularity")
plt.ylabel("Popular movie")
plt.show
```
