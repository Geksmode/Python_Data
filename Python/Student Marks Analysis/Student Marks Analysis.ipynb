```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df_marks = pd.read_csv('C:\\Users\\Geksmode\\Desktop\\data\\StudentsPerformance.csv')
```

```python
df_marks
```

```python
df_marks["gender"].value_counts().plot.pie(autopct='%1.1f%%')
plt.xlabel('gender')
plt.ylabel('Student')
```

```python
df_marks.describe()
```

```python
df_score = df_marks[['math score',"reading score","writing score"]].copy()
row_avg = df_score.mean(axis=1)
df_marks["average_grade"] = row_avg
df_marks
```

```python
df_marks.insert(6,"letter_grade_math","")
df_marks.insert(8,"letter_grade_reading","")
df_marks.insert(10,"letter_grade_writing","")

df_marks.insert(7,"Math_PassStatus","P")
df_marks.insert(9,"Writing_PassStatus","P")
df_marks.insert(11,"Reading_PassStatus","P")
```

```python
#allocate letter to marks for math

df_marks.loc[df_marks['math score'] < 60, 'letter_grade_math'] = "F"
df_marks.loc[df_marks['math score'] > 59 , 'letter_grade_math'] = "D"
df_marks.loc[df_marks['math score'] > 69 , 'letter_grade_math'] = "C"
df_marks.loc[df_marks['math score'] > 79  , 'letter_grade_math'] = "B"
df_marks.loc[df_marks['math score'] > 89, 'letter_grade_math'] = "A"
```

```python
#allocate letter to marks for reading

df_marks.loc[df_marks['reading score'] < 60, 'letter_grade_reading'] = "F"
df_marks.loc[df_marks['reading score'] > 59 , 'letter_grade_reading'] = "D"
df_marks.loc[df_marks['reading score'] > 69 , 'letter_grade_reading'] = "C"
df_marks.loc[df_marks['reading score'] > 79  , 'letter_grade_reading'] = "B"
df_marks.loc[df_marks['reading score'] > 89, 'letter_grade_reading'] = "A"
```

```python
#allocate letter to marks for writing

df_marks.loc[df_marks['writing score'] < 60, 'letter_grade_writing'] = "F"
df_marks.loc[df_marks['writing score'] > 59 , 'letter_grade_writing'] = "D"
df_marks.loc[df_marks['writing score'] > 69 , 'letter_grade_writing'] = "C"
df_marks.loc[df_marks['writing score'] > 79  , 'letter_grade_writing'] = "B"
df_marks.loc[df_marks['writing score'] > 89, 'letter_grade_writing'] = "A"
```

```python
#allocate letter to marks for average

df_marks.loc[df_marks['average_grade'] < 60, 'letter_grade_average'] = "F"
df_marks.loc[df_marks['average_grade'] > 59 , 'letter_grade_average'] = "D"
df_marks.loc[df_marks['average_grade'] > 69 , 'letter_grade_average'] = "C"
df_marks.loc[df_marks['average_grade'] > 79  , 'letter_grade_average'] = "B"
df_marks.loc[df_marks['average_grade'] > 89, 'letter_grade_average'] = "A"
```

```python
df_marks.loc[df_marks["letter_grade_math"]=="F","Math_PassStatus"] = "F"
df_marks.loc[df_marks["letter_grade_reading"]=="F","Reading_PassStatus"] = "F"
df_marks.loc[df_marks["letter_grade_writing"]=="F","Writing_PassStatus"] = "F"

df_marks
```

```python
df_marks["letter_grade_average"].value_counts().plot.pie(autopct='%1.1f%%')
plt.xlabel('Grade')
plt.ylabel('Student')
```

```python
p = sns.countplot(x='parental level of education', data = df_marks, hue='letter_grade_math', palette='bright')
_ = plt.setp(p.get_xticklabels(), rotation=90) 
```

```python
sns.countplot(x="letter_grade_average", data = df_marks, order=['A','B','C','D','F'],  palette="muted")
plt.show()
```

```python
fig, axes = plt.subplots(1,3, sharex = True, figsize = (10,5))
sns.countplot(ax = axes[0],x="letter_grade_math", data = df_marks, order=['A','B','C','D','F'],  palette="muted")
sns.countplot(ax = axes[1],x="letter_grade_reading", data = df_marks, order=['A','B','C','D','F'],  palette="muted")
sns.countplot(ax = axes[2],x="letter_grade_writing", data = df_marks, order=['A','B','C','D','F'],  palette="muted")
plt.show()
```

```python
fig, axes = plt.subplots(1,3, sharex = True, figsize = (10,5))
a = sns.countplot(ax = axes[0],x="parental level of education", data = df_marks, hue='Math_PassStatus',  palette="muted")
b = sns.countplot(ax = axes[1],x="parental level of education", data = df_marks,hue='Reading_PassStatus',  palette="muted")
c = sns.countplot(ax = axes[2],x="parental level of education", data = df_marks, hue='Writing_PassStatus',  palette="muted")
_ = plt.setp(a.get_xticklabels(), rotation=90)
_ = plt.setp(b.get_xticklabels(), rotation=90) 
_ = plt.setp(c.get_xticklabels(), rotation=90) 
```

```python
#transform data : object type to category type to procede to correlation

df_corr = df_marks[['gender',"race/ethnicity", "parental level of education","lunch","test preparation course","math score","writing score","reading score"]].copy()

df_corr['gender']=df_marks['gender'].astype('category').cat.codes
df_corr['race/ethnicity']=df_marks['race/ethnicity'].astype('category').cat.codes
df_corr['parental level of education']=df_marks['parental level of education'].astype('category').cat.codes
df_corr['lunch']=df_marks['lunch'].astype('category').cat.codes
df_corr['test preparation course']=df_marks['test preparation course'].astype('category').cat.codes
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_corr.corr(method = "pearson"),cmap='viridis',annot=True)
```
