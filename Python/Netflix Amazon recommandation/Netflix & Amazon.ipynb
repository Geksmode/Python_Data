```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
from ast import literal_eval
from sklearn.feature_extraction.text import CountVectorizer
df_netflix = pd.read_csv('C:\\Users\\Geksmode\\Desktop\\data\\netflix_titles.csv')
df_amazon = pd.read_csv('C:\\Users\\Geksmode\\Desktop\\data\\amazon_prime_titles.csv')
```

```python
df_series = pd.concat([df_netflix,df_amazon])
```

```python
df_series['Origine']= df_series['Origine'].fillna("Amazon_Prime")
df_series['cast']= df_series['cast'].fillna("")
df_series['director']= df_series['director'].fillna("")
df_series['country']= df_series['country'].fillna("")

df_series
```

```python
#content-based filter

tfidf = TfidfVectorizer(stop_words='english')
df_series['description']= df_series['description'].fillna("")
tfidf_matrix = tfidf.fit_transform(df_series["description"])

tfidf_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfidf_matrix,tfidf_matrix)
```

```python
indices = pd.Series(df_series.index,index=df_series['title']).drop_duplicates()
```

```python
def get_recommendations_description(title,cosine_sim=cosine_sim):
    idx=indices[title]
    #on récupère tous les films ressemblant au titre inscrit
    sim_scores = list(enumerate(cosine_sim[idx]))
    #Trier les films basé sur le score de similarité
    sim_scores = sorted(sim_scores,key=lambda x:x[1], reverse = True)
    sim_scores = sim_scores[1:11]
    series_indices = [i[0] for i in sim_scores]
    r_title = df_series["title"].iloc[series_indices]
    r_origine = df_series['Origine'].iloc[series_indices]
    frame = {'title': r_title,'Origine': r_origine}
    df_reco = pd.DataFrame(frame)
    
    return df_reco 

get_recommendations_description('Safe Haven')
```

```python
# Returns the list top 3 elements or entire list; whichever is more.
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['cast', 'director', 'listed_in']

for feature in features:
    df_series[feature] = df_series[feature].apply(clean_data)
df_series[features]
```

```python
def create_soup(x):
    return ''.join(x['cast']) + ' '.join(x['director']) + ' '.join(x['listed_in'])
df_series['soup'] = df_series.apply(create_soup, axis=1)
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df_series['soup'])
```

```python
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```
